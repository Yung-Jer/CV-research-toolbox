```python
import matplotlib.pyplot as plt
import numpy as np
from dncnn import *
```

```python
img = imread_uint('./random_img.jpg')
plt.imshow(img)
```

```python
# Since the difference between the 2D-Power Spectral Density 
# of different channels is significant, hence we use log-minmax
# scaling here
plt.imshow(log_normalize_image(psd2d(img)))
```

```python
# random crop and resize the image
img = random_crop_resize(img)
psd2d_img = psd2d(img)
spatial_frequency = np.arange(139) / 139

fig = plt.figure(figsize=(8, 5))
ax = fig.add_subplot(1, 1, 1)
colors = ['red', 'green', 'blue']
labels = ['Red', 'Green', 'Blue']
for idx, ele in enumerate([azimuthalAverage(psd2d_img[...,0]), azimuthalAverage(psd2d_img[...,1]), azimuthalAverage(psd2d_img[...,2])]):
    ax.semilogy(spatial_frequency, ele, color=colors[idx], alpha=0.5, label=labels[idx])
    ax.set_title('1D Radially Averaged Power Spectrum Density (PSD) of RGB channels', fontsize=8)
    ax.set_xlabel('Spatial Frequency Bins')
    ax.set_ylabel('Power Spectrum')
ax.legend()
plt.show()
```
